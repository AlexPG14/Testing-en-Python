```python
# Función básica
def suma(a, b):
    return a + b

# Test simple
def test_suma():
    assert suma(2, 3) == 5
```

## 📊 Ejemplo con datos

```python
import pandas as pd

# Función de limpieza
def limpiar(df):
    return df.dropna()

# Test
def test_limpiar():
    df = pd.DataFrame({"a": [1, None, 3]})
    result = limpiar(df)
    assert result.shape[0] == 2
```

## ⚠️ Edge Cases

```python
def calcular_media(lista):
    return sum(lista) / len(lista)

# Test edge case
def test_lista_vacia():
    try:
        calcular_media([])
    except ZeroDivisionError:
        assert True
```

## 🚀 Pytest avanzado

```python
import pytest

@pytest.mark.parametrize("input,output", [
    ([1,2,3], 2),
    ([10,20], 15),
])
def test_media_param(input, output):
    assert calcular_media(input) == output
```
